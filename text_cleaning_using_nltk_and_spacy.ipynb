```python
import nltk
import wikipedia
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string
```

```python
# Download necessary NLTK Data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# Get text
text = wikipedia.page('Geoffrey Hinton').content
```

```python
# Step -1: Lowercase
text = text.lower()
```

```python
# Step-2: Tokenize
tokens = word_tokenize(text)
```

```python
# Step-3 : Remove punctuations using NLTK's string.punctuation
tokens = [word for word in tokens if word not in string.punctuation]
```

```python
# Step-4 : Remove Stopwords:
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word not in stop_words and word.isalpha()]
```

```python
# Step-5 : Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]
```

```python
import spacy
```

```python
# Load spaCy English model
nlp = spacy.load("en_core_web_sm")

# Get text 
text = wikipedia.page("Geoffrey Hinton").content
```

```python
# STep 1-5 combined : spaCy handles lowercasing, tokenization, puctuaitons, topwords and lemmatization
doc = nlp(text.lower())
```

```python
# Fitered tokens : remove stopwords, punctuations and non-alphabetic tokens
filtered_tokens = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct and token.is_alpha]
```

```python
# Final cleaned text
cleaned_text_spacy = ' '.join(filtered_tokens)
print("===== SPACY CLEANED TEXT =====")
print(cleaned_text_spacy[:1000])  # show only first 1000 characters
```
