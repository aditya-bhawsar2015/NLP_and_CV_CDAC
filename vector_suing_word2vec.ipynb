```python
# Find vector using Word2Vec for each word from the pretrained model
import spacy

# Load English NLP Model
nlp = spacy.load('en_core_web_lg')
```

```python
# Example sentence
statement1 = "He deosited cash in the bank."
statement2 = "She sat by the river bank."

# Process the satatement with spaCy
doc1 = nlp(statement1)
doc2 = nlp(statement2)

# Access word vectors for each token in the processed sentences
print("Word vectors for statement 1 : ")
for token in doc1:
    print(f"{token.text} : {token.vector}")

print("Word vectors for sentence 2 : ")
for token in doc2:
    print(f"{token.text} : {token.vector}")
    
```
