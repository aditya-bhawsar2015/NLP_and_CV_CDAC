```python
import cv2
import matplotlib.pyplot as plt
```

```python
# load the image in color
image = cv2.imread('colorful-tiger.jpg')
```

```python
# Display image properties : 
print("Size: " , image.shape[:2])       # Height and Width
print("Mode: ","BGR")     # OpenCV loads images in BGR format by default
print("Format: JPEG")     # Format information usually comes from file type(if known)
```

```python
# Resize the image to maintain aspect ratio
width = 500               # Set desired width
height = int(image.shape[0] * (width / image.shape[1]))      # Maintain the aspect ratio
resized_image = cv2.resize(image, (width, height))
```

```python
# Convert the image to BGR for current display in Matplotlib
resized_image_rgb = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)
```

```python
# Show the resized image using Matplotlib
plt.imshow(resized_image_rgb)
plt.axis("off")     # Hide axes
plt.title("Colorful Tiger")
plt.show()
```

```python
# Convert the image to grayscale
gray_image = cv2.cvtColor(resized_image, cv2.COLOR_BGR2GRAY)
```

```python
# Save the new grayscale image
cv2.imwrite('colourful-tiger-gray.jpg', gray_image)
```
