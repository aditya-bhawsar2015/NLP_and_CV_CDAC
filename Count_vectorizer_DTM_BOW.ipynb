```python
# Use of CountVectorizer to create DTM/BOW
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
```

```python
# Sample documents
documents = [
    "I love natural language processing.",
    "Count Vectorizer is useful in NLP.",
    "NLP allows machines to understand human language."
]
```

```python
# Create a CountVectorizer instnce
vectorizer = CountVectorizer()

# Fit and transform the documents
X = vectorizer.fit_transform(documents)
```

```python
# Convert the matrix to dataframe for better visualization
columns = vectorizer.get_feature_names_out()
df = pd.DataFrame(X.toarray(), columns=columns)

print("Vocabulary : ", vectorizer.vocabulary_)       # shows the index of each word assigned by CountVectorizer
```
